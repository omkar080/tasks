```python
 import matplotlib.pyplot as plt
 import numpy as np
```

```python

years = [2001, 2002, 2003, 2004, 2005]

Shewag = [ 500, 800, 1100, 1300, 1500] 
Kohli = [300, 800, 1200, 1500, 1700]
Sachin = [500, 700, 1100, 1500, 1000]

x = np.arange(len(years))
arr = np.arange(len(years))
width = 0.25

plt.bar(x - width, Shewag, width=width, label="Shewag")
plt.bar(x, Sachin, width=width, label="Sachin")
plt.bar(x + width, Kohli, width=width, label="Kholi")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Comparison")
plt.xticks(arr, years)
plt.legend()
plt.tight_layout()
```

```python
OS = ["Android", "IOS", "Others"]
shares = [9000, 15000, 12000]

clrs = ['#ff9999', '#66b3ff', '#99ff99']

expld = [0.1,0.01,0.1]

plt.pie(shares, labels=OS, colors=clrs, explode=expld, shadow=True, startangle=100, autopct='%1.1f%%')
plt.title("Shares of different OS")
plt.show()
```
